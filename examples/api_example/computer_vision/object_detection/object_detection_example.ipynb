## Experiment on object detection

At the first step we need to import the necessary libraries and packages, and also configure a path to dataset.
The example uses a dataset that can be downloaded [here](https://www.kaggle.com/datasets/parohod/warp-waste-recycling-plant-dataset?select=Warp-D).

```python
import os

from fedot_ind.api.main import FedotIndustrial

DATASETS_PATH = os.path.abspath('Warp-D')
```

In this dataset, there is no file that describes the dataset in the YOLO format, so we will create it ourselves. You can read more about the YOLO format [here](https://docs.ultralytics.com/datasets/detect/).

```python
import yaml

with open(os.path.join(DATASETS_PATH, 'classes.txt'), 'r') as f:
    classes = f.read().splitlines()

config = {
    'train': 'train/images',
    'val': 'test/images',
    'nc': 28,
    'names': classes
}

with open(os.path.join(DATASETS_PATH, 'warp.yaml'), 'w') as f:
    yaml.dump(config, f)
```

As it was described in other examples, we need to instantiate the class FedotIndustrial with appropriate task type. Also, as the important parameter either the number of classes or torch model should be passed. Class 0 in the object detection task is reserved for the background, so we need to add 1 to the number of classes.

```python
fed = FedotIndustrial(task='object_detection', num_classes=29)
```

Next, we are able to start model training:

```python
fitted_model = fed.fit(
    dataset_path=os.path.join(DATASETS_PATH, 'warp.yaml'),
    dataset_name='WaRP',
)
```

Model predictions can be obtained in the following way:

```python
predict = fed.predict(data_path=os.path.join(DATASETS_PATH, 'test/images'))
predict_proba = fed.predict_proba(data_path=os.path.join(DATASETS_PATH, 'test/images'))
```

```python
print(predict)
```

```python
from fedot_ind.core.architecture.datasets.visualization import draw_sample_with_bboxes
import random

image = random.choice(list(predict.keys()))
fig = draw_sample_with_bboxes(image=image, target=predict[image], prediction=predict_proba[image], threshold=0.2)
```
